```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.basemap import Basemap
```

```python
fig=plt.figure(figsize=(8,8), dpi=100)
m = Basemap(projection = 'lcc',resolution='h',
          width=7E6, height=7E6,
          lat_0=21.43,lon_0=87.01)
m.etopo(scale=0.5, alpha=0.5)
m.bluemarble(scale=0.5)
#28.450978, 76.768688 location of our college

x,y = m(76.768688, 28.450978)
plt.plot(x,y, 'ok', markersize=6)
plt.text(x,y, 'My college', fontsize=13);
plt.savefig('mycollege.png', dpi=600, bbox_inches='tight')
plt.show()
```
